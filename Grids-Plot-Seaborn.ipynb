```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.species.unique()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g= sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
