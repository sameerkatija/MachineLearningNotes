### .loc and .iloc 

```python
animals = pd.Series(["cat", "dog", "bird","panda","snake"])
```

```python
# return the data on index 3
animals.loc[3]
```

```python
# return the data on position 3
animals.iloc[3]
```

### As these both seems same but let's change the index and duplicate the index now to see the actual difference

```python
animals = pd.Series(["cat", "dog", "bird","panda","snake"],index=[1,2,3,4,3])
```

```python
# As you can see i have change the index and there are index 3 duplicated
animals
```

```python
# This time loc will return all the element on index 3
animals.loc[3]
```

```python
# While the iloc will return the element on position 3
animals.iloc[3]
```

## Manipulating Data

```python
car_sales = pd.read_csv('https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/exported-car-sales.csv', index_col=False)
```

```python
car_sales
```

```python
### Deleting a column 
del car_sales['Unnamed: 0']
```

```python
car_sales
```

```python
car_sales.dtypes
```

```python
### Converting the price to int from object
```

```python
car_sales['Price'] 
```

```python
# removing special symbols.
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,]', '',regex=True)
```

### I could have done the formatting in one part but i chose to seperate it to help you to understand

```python
# removing . and value after dot 
car_sales['Price'] = car_sales['Price'].str.replace('\.[^\.]\S$', '',regex=True)
```

### Note: In replace function I have used incase you don't know about regex, you can visit python offical document and read here about [regex](https://docs.python.org/3/howto/regex.html)

**To play around with regex use this [link](https://regexr.com/)**

```python
# Let's check the datatype of car_sales['price']
car_sales.dtypes
```

```python
# Converting Object to python 
car_sales['Price'] = car_sales['Price'].astype(int) 
```

```python
# Let's check the datatype of car_sales['price']
car_sales.dtypes
```

```python
### Converting the str in make column to lowercase
car_sales['Make'] = car_sales['Make'].str.lower()
```

```python
car_sales
```

### Let's work with Missing Data (As In Real world we might encounter this scenario)

```python
car_sales_missing = pd.read_csv('https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-missing-data.csv')
```

```python
# Note here Nan (Not a numer) showing us that this is empty
car_sales_missing
```

### Let's try the mean of Odometer column's missing values (note: this may not be the case in real world)
#### This is just to introduce you with .fillna function and a possibility 

```python
#Inplace is same as car[odo] = car[odo].fillna(....)
car_sales_missing['Odometer'].fillna(car_sales_missing['Odometer'].mean(), inplace=True)
```

```python
car_sales_missing
```

### Let's say we don't want to fill other columns with mean like doors so in this case we have to drop those rows

```python
car_sales_missing.dropna(inplace=True)
```

```python
car_sales_missing
```
