## Two ways of plotting

### 1s't method (Functional Method)

```python
plt.plot(x,y);
plt.xlabel('X Label')
plt.ylabel('Y label')
plt.title('Title');
```

### Second mehtod (Object Oriented Method)

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title');
```

## Matplotlib workflow

### 1. Importing matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

### 2. Preparing data

```python
x = [1,2,3,4]
y = [11,22,33,44]
```

### 3. Setup plot

```python
fig,ax = plt.subplots(figsize=(10,10))
```

### 4. Plot data

```python
ax.plot(x,y)
```

### 5. Customize plot

```python
ax.set(title='Simple Plot', xlabel='x-axis',ylabel='y-axis');
```

### 6. Show and save the image

```python
fig.savefig('./plot.png')
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set(title='Simple Plot', xlabel='x-axis',ylabel='y-axis');
plt.show()
```

### There are two ways for plotting as shown above it's totally personal preference to choose one, I basically love function way but OO plotting is recommended by many professionals.

### Creating subplots (I will use both functional and OO way of plotting interchangingly)

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x)
plt.show()
```

### Creating plot using oo plotting indise to one plot

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.53,0.4,0.3])
x1 = np.linspace(0,5,15)
y1 = x1 ** 2
axes1.plot(x1,y1)
axes1.set_title('Larger Plot')
axes2.plot(y1,x1)
axes2.set_title('Smaller Plot')
```

### Creating plot using oo plotting side by side

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.6,1])
axes2 = fig.add_axes([0.8,0.1,0.6,1])
x1 = np.linspace(0,5,15)
y1 = x1 ** 2
axes1.plot(x1,y1)
axes1.set_title('Larger Plot')
axes2.plot(y1,x1)
axes2.set_title('Smaller Plot')
```

### Creating a subplot using Object Oriented (OO) plotting method

```python
fig , axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x1,y1)
axes[0].set_title('First Plot')
axes[1].plot(y1,x1)
axes[1].set_title('Second Plot')
plt.tight_layout()
```

### Creating multiple plots on same figure

```python
plt.plot(x1, x1 ** 2)
plt.plot(x1,x1**3)
plt.ylim(0,130)
plt.xlim(0,5)
plt.title('Multiple Plots')
plt.xlabel('X Label')
plt.ylabel('Y label')
```

### Creating a legend

```python
plt.ylim(0,130)
plt.xlim(0,5)
plt.title('Multiple Plots')
plt.xlabel('X Label')
plt.ylabel('Y label')
plt.plot(x1, x1 ** 2, label='X square')
plt.plot(x1,x1**3, label='X cube')
plt.legend()
```

## Customizing Plots

### Changing the color of line

**Note: There are many colors available in [mtplotlib](https://matplotlib.org/stable/gallery/color/named_colors.html). If you want to add cutom color you can also use hex code** 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='lightgreen');
```

### Adding the width to line

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orchid', linewidth=3);
```

### Line Style 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,c='orchid', lw=3 , alpha=0.9, linestyle=':');
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x1,y1,c='orchid', lw=3 , alpha=0.9, linestyle='--');
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x1,y1,c='orchid', lw=3 , alpha=0.6, linestyle='-', marker='*', markerfacecolor='r', markersize=20);
```

```python
plt.scatter(x1,y1,marker='+');
```
